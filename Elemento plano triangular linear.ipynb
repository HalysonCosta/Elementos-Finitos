```python
def area_elemento_triangular(no1,no2,no3):
    a = np.sqrt((no1[0]-no2[0])**2 + (no1[1]-no2[1])**2)
    b = np.sqrt((no2[0]-no3[0])**2 + (no2[1]-no3[1])**2)
    c = np.sqrt((no3[0]-no1[0])**2 + (no3[1]-no1[1])**2)

    semperimetro = (a+b+c)/2
    area = np.sqrt(semperimetro*(semperimetro-a)*(semperimetro-b)*(semperimetro-c))

    return area

def arestas_elemento_triangular(no1,no2,no3):
    a = np.sqrt((no1[0]-no2[0])**2 + (no1[1]-no2[1])**2)
    b = np.sqrt((no2[0]-no3[0])**2 + (no2[1]-no3[1])**2)
    c = np.sqrt((no3[0]-no1[0])**2 + (no3[1]-no1[1])**2)

    return a,b,c
```
